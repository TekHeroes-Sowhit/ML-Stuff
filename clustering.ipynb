```python
df=pd.read_csv('file:///C:/Users/00824732/Desktop/Classification-data.csv')
```

```python
df.columns
```

```python
plt.scatter(df['Age'],df['Income'])
plt.show()
```

```python
km=KMeans(n_clusters=3)
km
```

```python
df.columns
```

```python
y_predicted=km.fit_predict(df[['Age','Income']])
```

```python
df['Cluster']=y_predicted
```

```python
df0=df[df.Cluster==0]
df1=df[df.Cluster==1]
df2=df[df.Cluster==2]
plt.scatter(df0.Age,df0['Income'],color="green")
plt.scatter(df2.Age,df2['Income'],color="green")
plt.scatter(df2.Age,df2['Income'],color="green")
plt.show()
```

```python
Max=MinMaxScaler()
Max.fit(df[['Income']])
df['Income']=Max.transform(df[['Income']]) 
Max.fit(df[['Age']])
df.Age=Max.transform(df[['Age']])
# df.Age=Max.transform(df[[Age]])
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['Age','Income']])
y_predicted
df['cluter']=y_predicted
df.drop('cluter',axis='columns',inplace=True)
```

```python
km.cluster_centers_
```

```python
df0=df[df.Cluster==0]
df1=df[df.Cluster==1]
df2=df[df.Cluster==2]
plt.scatter(df0.Age,df0['Income'],color="green")
plt.scatter(df2.Age,df2['Income'],color="blue")
plt.scatter(df2.Age,df2['Income'],color="yellow")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='red')
plt.show()
```

```python
## Elbow plot method
im = Image.open(r"C:/Users/00824732/Desktop/k_means_2.PNG") 
im
```

```python
k_rng=range(1,3)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income']])
    sse.append(km.inertia_)
plt.plot(k_rng,sse)
plt.show()
```
